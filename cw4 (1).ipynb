```python
import random
random.seed(150509)
randomlist = []
for a in range(200):
  randomlist.append(random.randint(0,100))
zad1_suma = 0
zad2_minimum =randomlist[0]
zad3_maximum =randomlist[0]
for x in randomlist:
  zad1_suma += x
  if x <= zad2_minimum:
    zad2_minimum = x
  if x >= zad3_maximum:
    zad3_maximum = x
print(f'zad1_suma = {zad1_suma}')

#-------------------------------------------

print(f'zad2 Najmniejsza wbudowana f {min(randomlist)}')
print(f'zad2 Najmniejsza normalnie {zad2_minimum}')

#-------------------------------------------

print(f'zad3 Najwieksza wbudowana f {max(randomlist)}')
print(f'zad3 Najwieksza normalnie {zad3_maximum}')

#-------------------------------------------

posortowana = sorted(randomlist.copy())
print(f'zad4 mediana - {posortowana[100]}')

#-------------------------------------------

zad5 = randomlist.copy()
zad5_lista1 = sorted(zad5)
zad5_lista2 = zad5
print('zad5')
for q in range(20):
  print(zad5_lista1[q], end=' ')
print('\n')
for w in range(len(zad5_lista2)):
  for e in range(len(zad5_lista2)-1):
    if zad5_lista2[e] > zad5_lista2[e+1]:
      pom = zad5_lista2[e]
      zad5_lista2[e] = zad5_lista2[e+1]
      zad5_lista2[e+1] = pom
for r in range(20):
  print(zad5_lista2[r], end=' ')
print('\n')

#-------------------------------------------

zad6 = 1
for t in randomlist:
  zad6 *= t
print(f'zad6 - {zad6}')

#-------------------------------------------

zad7 = 0
for y in randomlist:
  if y>=100:
    zad7+=1
print(f'zad7 - liczba liczb 3 cyfrowych {zad7}')

#-------------------------------------------
 
zad8 = 0
zad8_cyfra = 0
for u in randomlist:
  if randomlist.count(u)> zad8:
    zad8 = randomlist.count(u)
    zad8_cyfra = u

print(f'zad 8  - liczba {zad8_cyfra} wystepuje - {zad8}x')

#-------------------------------------------

print('zad9 - unikalne wartosci ', end=' ')
for i in randomlist:
  if randomlist.count(i)==1:
    print(i, end=' ')
print('\n')
print('zad10 - wartosci powtarzajace sie 3 razy ', end=' ')
zad10 = []
for i in randomlist:
  if randomlist.count(i)==3:
    if i not in zad10:
      zad10.append(i)
for i in zad10:
  print(i, end=' ')
print('\n')

#-------------------------------------------

zad11 = 0
for o in randomlist:
  if o >= 27:
    zad11 += 1
print(f'zad11 ilosc >27- {zad11}')

#-------------------------------------------

print(f'zad 12 : 3 najmniejsze {posortowana[0:3]}')
print(f'zad 12 : 3 najwieksze {posortowana[-3:]}')

#-------------------------------------------

zad13 = randomlist.copy()
unikaty = []
for p in zad13:
  if p not in unikaty:
    unikaty.append(p)
print(f'zad 13 ilosc liczb bez dubliaktow - {len(unikaty)}')

#-------------------------------------------

print('zad14')
dict = {}
for a in range(1,101):
  dict[a] = 0
for s in randomlist:
  dict[s]+=1
for k,v in dict.items():
  print(f'{k} - {v}')

#-------------------------------------------

dane = input().split(' ')
a, b = int(dane[0]), int(dane[1])
zad15 = 0
for d in randomlist:
  if d>= a and d<=b:
    zad15 += 1
print(f'zad15 - w przedziale wystepuje {zad15} liczb')

#-------------------------------------------

zad16 = []
zad17 = []
for f in randomlist:
  if f%2==0:
    zad16.append(f)
  if f%2==1:
    zad17.append(f)

print(f'zad 16 - ilosc parzystych {len(zad16)}')
print(f'zad 17 - ilosc nieparzystych {len(zad17)}')

zad16 = sorted(zad16)
zad17 = sorted(zad17, reverse=True)

print(f'zad 18 - {zad16+zad17}')




```
